##### The problem with modeling long sequences

**Problem**

    The problem with pre-LLM architectures that do not include attention mechanisms

* Suppose we want to develop a `language translation model that translates text from one language into another`
    * We cannot simply translate a text word by word due to the grammatical strucutres in the source and target language. e.g., see image below

    


* To address this problem, it is common to use a `deep neural network with two submodules`
    * Encoder
    * Decoder
* `Encoder:` The job of encoder is to first read in and process the entire text.
* `Decoder:` Then decoder produces the translational text.

Before the advent of transformers, `RNNs` were the most popular `encoder-decoder architecture for language translation`

* An RNN is type of nn where `outputs from previous step are fed as inputs to the current step`
* _Making them well suited for sequential data like text_

**Encoder-Decoder RNN**
* The input-text is fed into the encoder
    * which processes it sequentially
    * The encoder updates it hidden state (the internal values at the hidden layers) at each step _trying to capture the entire meaning of the input sentence in the final hidden state see in the image below_
* _The decoder then takes this hidden state to start generating the translated sentence, **one word at a time**_
    * It also updates its hiddent state at each step, which is supposed to carry the context necessary for the next-word prediction

    


**Summarized Notes (problem with seq2seq modeling)**
* RNN earlier encoder-decoder architecter before transformers
* Encoder:
    * Encoder process the words sequentially updates its hidden state
    * Final word(hidden state) contains the full meaning of the sentence
* Decoder:
    * Taken the hidden state from encoder
    * `Predicts one word at a time`
    * And also updates the hidden state to `carry forward the context necessary for next-word prediction`

_works well for the short sentences_

The **entire input sentence is compressed into a single vector(the final hidden state)**
**For long sentences:**
* `Information can be lost`
* `Early words may fade (vanishing gradient problem)`
* `Long-range dependicies becomes hard to capture`
    * Don't work well for longer texts `as thex dont have direct access to previous words in the input`
* **`Major shortcoming in this approach is the RNN must remeber the entire encoded input in a single (final) hidden state before passing it to the decoder see the image below`**

_This leads to the attention mechanism_

### Capturing data dependices with attention mechanism

* Don't work well for longer texts `as thex dont have direct access to previous words in the input`
* **`Major shortcoming in this approach is the RNN must remeber the entire encoded input in a single (final) hidden state before passing it to the decoder see the image above (german to english translation one)`**

__Hence researchers developed the _Bahdanau attention mechanism for RNNs in 2014___
    * This modifies the entire  encoder-decoder RNN such that `decoder can selectively access different parts of the input sequence at each decoding steps as shown in the image below.`
         
    
Interestingly, after 3 years researchers found that RNN architectures are not required for building a deep neural networks for NLP and proposed the **original transformer architecture _including a self-attention mechanism inspired by the Bahdanau attention mechanism_**
